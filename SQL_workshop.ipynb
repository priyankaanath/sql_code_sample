```python
%%sql
create table project(
project_id integer(6) primary key, 
project_name text(20), 
project_owner integer(6),
customer_id integer(6),
technology_id integer(6),
domain text(20),
team_size integer(6),
contract_value integer(10),
foreign key (customer_id) references customer(customer_id),  
foreign key (technology_id) references technology(technology_id));
```
